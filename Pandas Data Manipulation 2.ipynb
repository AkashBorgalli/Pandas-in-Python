```python
left = pd.DataFrame({ 'Key':['K0','K8','K2','K3'],
    'A':['A0','A1',"A2","A3"],
    'B':['B0','B1',"B2","B3"]
    
})
right = pd.DataFrame({ 'Key':['K0','K1','K2','K3'],
    'C':['C0','C1',"C2","C3"],
    'D':['D0','D1',"D2","D3"]
    
})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on='Key')
```

```python
## left join 
## meaning from left table keys its will try to match with right and we it get that key then it will merge
## you will see in table that there are left table keys in merge table
pd.merge(left,right,on='Key',how='left')
```

```python
## right join 
## meaning from right table keys its will try to match with left and we it get that key then it will merge
## you will see in table that there are right table keys in merge table
pd.merge(left,right,on='Key',how='right')
```

```python
## outer join
## meaning it will mearge from both side
pd.merge(left,right,on='Key',how='outer')
```
